# Python Basics

## 1. Importing Libraries

```python
import numpy as np  # Import NumPy for numerical operations
import matplotlib.pyplot as plt  # Import Matplotlib for plotting
```

## 2. Variables

```python
x = 5.  # float variable
print(x)
```

## 3. NumPy Arrays

### 3.1 Creating arrays

```python
a = np.array([[1.], [2.], [3.]])       # 2D array, shape (3,1)
b = np.array([0., 3., 4.])             # 1D array, shape (3,)
c = np.array([[1, 13, -6],[9,4,7],[19,16,2]])  # 2D array, shape (3,3)
print('a =', a)
print('b =', b)
print('c =', c)
```

### 3.2 Array Shapes

```python
print('Shape of a:', a.shape)
print('Shape of b:', b.shape)
print('Shape of c:', c.shape)
```

### 3.3 Indexing

```python
print('0th row of c:', c[0,:])
print('2nd column of c:', c[:,2])
print('0th and 1st columns of c:', c[:,[0,1]])
```

## 4. Conditional Array Operations

```python
c[np.where(c<0)] = -1
print('c after replacing negatives:', c)
```

## 5. Creating Special Arrays

```python
print('np.arange(3):', np.arange(3))
print('np.arange(3,7,2):', np.arange(3,7,2))
print('np.zeros((2,2)):\n', np.zeros((2,2)))
print('np.empty((5,1)):\n', np.empty((5,1)))
```

## 6. Array Operations

```python
B = np.zeros(2)
A = np.zeros((2,2))
t = np.dot(B,A)
print('Dot product of B and A:', t)
```

## 7. Looping and Array Assignment

```python
p = np.empty((5,1))
for i in range(5):
    p[i] = i**2
print('Array of squares:', p)
```

## 8. Finding Array Properties

```python
pMax = np.max(p)  # maximum value
print('pmax=', pMax)
# np.max(array) returns the largest element in the array
```

## 9. While Loop

```python
i = 1
while i<6:
    print(i)
    i+=1
```

## 10. User Input and Conditional Statements

```python
nstr = input('Type an integer: ')
number = int(nstr)
if number>0:
    print('Positive number')
elif number==0:
    print('Zero')
else:
    print('Negative number')
```

## 11. String Formatting

```python
txt = "My name is {fname}, I'm age {age}".format(fname='John Doe', age='20')
print(txt)
```

## 12. Matplotlib Plotting Examples

```python
# Line plot
xvals = np.arange(-1,10,0.1)
yvals = np.sin(xvals)
plt.plot(xvals,yvals,'r--')
plt.title('Example plots')
plt.xlabel('Input')
plt.ylabel('Function values')
plt.savefig('fig1.png', dpi=150)
plt.show()
```

```python
# Line plot with legend
plt.plot(xvals,yvals,'r--',label='xvals')
plt.title('Example plots')
plt.xlabel('Input')
plt.ylabel('Function values')
plt.legend()
plt.show()
```

```python
# Scatter plot
plt.scatter(xvals,yvals,color='green',marker='<')
plt.xlabel('Input')
plt.ylabel('Function values')
plt.title('Scatter plot')
plt.show()
```

```python
# Single subplot
x = np.linspace(0,2*np.pi,400)
y = np.sin(x**2)
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Simple plot')
fig.suptitle('The suptitle text')
plt.show()
```

```python
# Two subplots
f,(ax1,ax2)=plt.subplots(1,2,sharey=True)
ax1.plot(x,y)
ax2.scatter(x,y)
ax1.set_title('Sharing Y axis')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.set_xlabel('x')
plt.show()
```
